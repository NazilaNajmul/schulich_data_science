```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading df in the enviorment
df = pd.read_csv("/Users/nazil/Downloads/Assignment1_StreamFlow_Subscription_Data.csv")
```

```python
#details of the original df
df.describe()
df.head()
```

```python
import datetime as dt
```

```python
#changing datatypes for Gender
df['Gender']=df["Gender"].astype('category')
```

```python
#changing datatypes for Location
df['Location']=df['Location'].astype('category')
```

```python
#changing datatype to datetime
df['Subscription_Start_Date'] = pd.to_datetime(df['Subscription_Start_Date'])
```

```python
#changing datatype to datetime
df['Date_of_Churn'] = pd.to_datetime(df['Date_of_Churn'])
```

```python
#updated df info to check the changed datatypes
df.info()
```

```python
#defining bins for the age group and adding it into the dataframe
bins= [19, 31, 41, 51, 61, 71]
df['age_group'] = pd.cut(df['Age'], bins=bins, right=False)
```

```python
#calculating duration of the subscribers
df["Duration"] = df['Date_of_Churn'] - df['Subscription_Start_Date']
```

```python
df.head()
```

```python
#slicing the dataframe based on churned and non-churned
non_churned_df = df.loc[df['Churned'] == 0]
```

```python
non_churned_df.describe()
```

```python
#slicing the dataframe based on churned and non-churned
churned_df = df.loc[df['Churned'] == 1]
```

```python
churned_df.head()
```

```python
churned_df.describe()
```

```python
#share of churned people location wise
location_count = churned_df.groupby('Location').size()
```

```python
print(location_count/location_count.sum()*100)
```

```python
#share of churned people age_group wise
age_group_count = churned_df.groupby('age_group').size()
```

```python
print(age_group_count/age_group_count.sum()*100)
```

```python
#percentage of each genre as favorite genre among churned
favorite_genre_count = churned_df.groupby('Favorite_Genre').size()
print(favorite_genre_count/favorite_genre_count.sum()*100)
```

```python
#reasons for churned of each genre as favorite genre among churned
reason_churned_count = churned_df.groupby('Reason_for_Churn').size()
print(reason_churned_count/reason_churned_count.sum()*100)
```

```python
#slicing the dataframe based on Reason, dissatisfaction with content
content_churned_df = churned_df.loc[df['Reason_for_Churn'] == 'Dissatisfied with content']
content_churned_df.head()
```

```python
content_churned_df.describe()
```

```python
#share of every genre as favorite genre among dissastified with content churn
content_dissastified_churned_count = content_churned_df.groupby('Favorite_Genre').size()
print(content_dissastified_churned_count/content_dissastified_churned_count.sum()*100)
```

```python
#age group wise genre preference
genre_age_wise = content_churned_df.groupby(['age_group', 'Favorite_Genre']).size()
```

```python
print(genre_age_wise/genre_age_wise.sum()*100)
```

```python
#slicing the dataframe based on Reason anything else other than dissatisfaction with content
non_content_churned_df = churned_df.loc[df['Reason_for_Churn'] != 'Dissatisfied with content']
non_content_churned_df.head()
```

```python
#genre preference among those who churned for other reasons
non_content_dissastified_churned_count = non_content_churned_df.groupby('Favorite_Genre').size()
print(non_content_dissastified_churned_count/non_content_dissastified_churned_count.sum()*100)
```

```python
#percentage of rating among churned
rating_given_count = churned_df.groupby('Rating_Given').size()
```

```python
print(rating_given_count/rating_given_count.sum()*100)
```

```python
from scipy.stats.mstats import winsorize
```

```python
#mean duration of the churned to see if there is any pattern
#applying winsorization to the 'Duration' column and create 'duration_winsorized' column to exclude extreme values
churned_df['duration_winsorized'] = winsorize(churned_df['Duration'], limits=[0.05, 0.05])

#calculating the mean of 'duration_winsorized' column
mean_winsorized = churned_df['duration_winsorized'].mean()
```

```python
print("Mean of duration_winsorized column:", mean_winsorized)
```
